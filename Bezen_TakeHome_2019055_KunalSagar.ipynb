### Importing Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Reading the Data. 
#### Please put the .csv file in the same directory as of the notebook, and rename it as 'data.csv' (or change the name of the file in the below cell

```python
df = pd.read_csv("data.csv")
```

### Using isnull() to check for the products without prices

```python
products_without_prices = df.loc[df["price_string"].isnull()]
products_without_prices
```

### Couting the uuid from dataframe to get the count for products without prices

```python
count_products_without_prices = len(products_without_prices.uuid)
count_products_without_prices
```

### Using notna() to check for not null values in price_string column

```python
products_with_prices = df.loc[df["price_string"].notna()]
products_with_prices
```

### Counting in total priced products in a similar manner

```python
count_products_with_prices = len(products_with_prices.uuid)
count_products_with_prices
```

### Counting Priced Products for each Product Type

```python
count_products_with_prices_product_type = products_with_prices.groupby(["product_type"])[["level_1", "category"]]
count_products_with_prices_product_type.count()
```

### Counting Priced Products for each Level 1

```python
count_products_with_prices_level_1 = products_with_prices.groupby(["level_1"])[["product_type", "category"]]
count_products_with_prices_level_1.count()
```

### Counting Priced Products for each Category

```python
count_products_with_prices_category = products_with_prices.groupby(["category"])[["product_type", "level_1"]]
count_products_with_prices_category.count()
```

### Replacing the 'dollar' sign to create uniformity in data
### Also changing the datatype from string to float

```python
products_with_prices["price_string"] = products_with_prices["price_string"].str.replace('$', '').astype(float)
products_with_prices
```

### Calculating the mean price for each category

```python
new_df = products_with_prices.groupby(["category"])[["price_string"]].mean()
new_df
```

### Plotting a bar graph for the average prices

```python
new_df.plot(y='price_string', kind = 'bar')
plt.legend(fontsize=8)
plt.show()
```

### Plotting a pie chart for the average prices

```python
new_df.plot.pie(y = 'price_string', figsize=(10,10), autopct='%.2f%%', startangle=90)
plt.legend(fontsize=6)
plt.show()
```
